# Auto differentiation:  

```python
import numpy as np
import matplotlib.pyplot as plt
from var import Variable
```

### Test with $Cos$ and $sin$ functions : 

First let $a \in \mathbb{R}$ so, 

$$ c = cos(a) \Longrightarrow \frac{\partial c}{\partial a} =  \frac{\partial}{\partial a} cos(a)$$

so, 

$$ \frac{\partial c}{\partial a} = sin(a) $$

```python
t = np.linspace(0, 10, 100)
```

```python
a = Variable(t)
```

```python
c = a.cos()
```

```python
c.backward()
```

```python
plt.plot(t, a.grad)
plt.plot(t, c.data)
```

### Test with $log$ function : 

First let $t \in \mathbb{R}^{+*}$ so, 

$$ c = log(a) \Longrightarrow \frac{\partial c}{\partial a} =  \frac{\partial}{\partial a} log(a)$$

so, 

$$ \frac{\partial c}{\partial a} = \frac{1}{a} $$

```python
t = np.linspace(0.5, 5, 100)
```

```python
a = Variable(t)
```

```python
c = a.log()
```

```python
c.backward()
```

```python
plt.plot(t, a.grad)
plt.plot(t, c.data)
```

### Test with quadratic function: 

First let $t \in \mathbb{R}^{+*}$ so, 

$$ c = a^2 \Longrightarrow \frac{\partial c}{\partial a} =  \frac{\partial}{\partial a} a^2$$

so, 

$$ \frac{\partial c}{\partial a} = 2a $$

```python
t = np.linspace(-3, 3, 100)
```

```python
a = Variable(t)
```

```python
c = a*a
```

```python
c.backward()
```

```python
plt.plot(t, a.grad)
plt.plot(t, c.data)
```

### Test with $\sqrt{x}$ function:

First let $t \in \mathbb{R}^{+*}$ so, 

$$ c = \sqrt{a} \Longrightarrow \frac{\partial c}{\partial a} =  \frac{\partial}{\partial a} \sqrt{a}$$

so, 

$$ \frac{\partial c}{\partial a} = \frac{1}{2\sqrt{a}} $$

```python
t = np.linspace(0.5, 100, 100)
```

```python
a = Variable(t)
```

```python
c = a.sqrt()
```

```python
c.backward()
```

```python
plt.plot(t, a.grad)
plt.plot(t, c.data)
```
